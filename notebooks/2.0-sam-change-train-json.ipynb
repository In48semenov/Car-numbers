```python
with open('train.json') as file:
    json_file = json.load(file)

df = pd.DataFrame({
    'path': [],
    'box': [],
    'text': []
})
    
for line in json_file:
    df_curr = pd.DataFrame({
        'path': [line['file'] for _ in range(len(line['nums']))],
        'box': [l['box'] for l in line['nums']],
        'text': [l['text'] for l in line['nums']]
    })
    df = df.append(df_curr, ignore_index=True)
```
